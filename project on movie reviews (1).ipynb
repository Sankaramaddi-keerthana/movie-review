```python
train_size = int(len(dataset)*.7)
train_texts = dataset['text'][:train_size]
train_tags = dataset['tag'][:train_size]

test_texts = dataset['text'][train_size:]
test_tags = dataset['tag'][train_size:]
```

```python
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Dense
from keras.preprocessing.text import Tokenizer
```

```python
max_words = 5000
tokenize = Tokenizer(num_words=max_words, char_level=False)
tokenize.fit_on_texts(train_texts) # only fit on train

x_train = tokenize.texts_to_matrix(train_texts)
x_test = tokenize.texts_to_matrix(test_texts)
```

```python
x_train
```

```python
x_test
```

```python
encoder = LabelEncoder()
encoder.fit(train_tags)
y_train = encoder.transform(train_tags)
y_test = encoder.transform(test_tags)
```

```python
y_test
```

```python
encoder.classes_
```

```python
import numpy as np
```

```python
import numpy as np
np.unique(y_train)
```

```python
from keras import utils
y_train_new = utils.to_categorical(y_train)
y_test_new =  utils.to_categorical(y_test)
```

```python
y_test_new
```

```python
y_train_new
```

```python
model = Sequential()
model.add(Dense(560, input_shape=(max_words,),activation='relu'))
```

```python
model.add(Dense(5,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train_new,batch_size=32,epochs=1,verbose=1,validation_split=0.1)
```
